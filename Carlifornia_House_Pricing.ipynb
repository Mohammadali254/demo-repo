```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = ('/content/drive/MyDrive/Carlifornia_housing.csv')
data = pd.read_csv(file_path)

data
```

```python
data.info
```

```python
data.dropna(inplace= True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

# selecting our features X and and predictor y

X = data.drop(['median_house_value'], axis= 1)
y = data['median_house_value']
```

```python
# splittig our data in to 80% training and 20% testing i.e test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# joining the training data
train_data = X_train.join(y_train)
train_data
```

```python
# Data Exploration
train_data.hist(figsize = (15,8))
```

```python
# Correlation metrics
train_data.corr()
```

```python
# Setting up a correlation heatmap
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
# making a normal curiving distribution
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)

train_data.hist(figsize=(15,8))
```

```python
# Getting the Ocean Proximity since it's a String data and we need to work with it
# train_data.ocean_proximity

# Performing the hot encoding to convert it to Binary
# pd.get_dummies(train_data.ocean_proximity)

# joining it in to our training data with dropping the un binary
# train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)

# saving the value in to a varibale again.
# Assuming the DataFrame containing the "ocean_proximity" column is called "data"

train_data = data.join(pd.get_dummies(data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)
```

```python
# Setting up a correlation heatmap
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
plt.figure(figsize = (15, 8))
sns.scatterplot(x = "latitude", y= "longitude", data = train_data, hue = "median_house_value", palette = "coolwarm")
```

```python
# performing some feature engeneering
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['households_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
# importing Linear regression
from sklearn.linear_model import LinearRegression
X_train, y_train = train_data.drop(['median_house_value'], axis=1),train_data['median_house_value']

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# doing the same for testing data.
test_data = X_test.join(y_test)
test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']+1)
test_data['population'] = np.log(test_data['population']+1)
test_data['households'] = np.log(test_data['households']+1)


test_data = data.join(pd.get_dummies(data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['households_rooms'] = test_data['total_rooms'] / test_data['households']


```

```python
X_test, y_test = test_data.drop(['median_house_value'], axis=1),test_data['median_house_value']
```

```python
reg.score(X_test, y_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```
